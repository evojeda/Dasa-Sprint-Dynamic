```python
#Estruturas de Busca

def busca_linear(lista, alvo):
    i = 0
    while i < len(lista):
        if lista[i] == alvo:
            return i
        i += 1
    return -1

insumos = ["Reagente A", "Reagente B", "Tubo 5ml", "Luva latex", "Álcool 70%"]
alvo = "Reagente B"

resultado = busca_linear(insumos, alvo)

if resultado != -1:
    print(f"O insumo '{alvo}' foi encontrado no índice {resultado}")
else:
    print("Insumo não encontrado")

def busca_binaria(lista, alvo):
    esquerda, direita = 0, len(lista) - 1
    while esquerda <= direita:
        meio = (esquerda + direita) // 2
        if lista[meio] == alvo:
            return meio
        elif lista[meio] < alvo:
            esquerda = meio + 1
        else:
            direita = meio - 1
    return -1

insumos = ["Álcool 70%", "Luva latex", "Reagente A", "Reagente B", "Tubo 5ml"]
alvo = "Reagente A"

resultado = busca_binaria(insumos, alvo)

if resultado != -1:
    print(f"O insumo '{alvo}' foi encontrado no índice {resultado}")
else:
    print("Insumo não encontrado")

```

```python
#Ordenação (Merge Sort e Quick Sort)
from datetime import datetime

insumos = []
n = int(input("Quantos insumos deseja cadastrar? "))
for i in range(n):
    print(f"\nInsumo {i+1}/{n}")
    nome = input("Nome do insumo: ")
    quantidade = int(input("Quantidade consumida: "))
    validade_raw = input("Data de validade (AAAA-MM-DD): ")
    insumos.append({"nome": nome, "quantidade": quantidade, "validade": validade_raw})

def merge(esq, dir, key):
    i = j = 0
    out = []
    while i < len(esq) and j < len(dir):
        if key(esq[i]) <= key(dir[j]):
            out.append(esq[i]); i += 1
        else:
            out.append(dir[j]); j += 1
    out.extend(esq[i:]); out.extend(dir[j:])
    return out

def merge_sort(lst, key):
    if len(lst) < 2:
        return lst[:]
    meio = len(lst) // 2
    esquerda = merge_sort(lst[:meio], key)
    direita  = merge_sort(lst[meio:], key)
    return merge(esquerda, direita, key)

def quick_sort(arr, key):
    if len(arr) <= 1:
        return arr[:]
    pivo = key(arr[-1])
    menores = [x for x in arr[:-1] if key(x) <= pivo]
    maiores = [x for x in arr[:-1] if key(x) >  pivo]
    return quick_sort(menores, key) + [arr[-1]] + quick_sort(maiores, key)

# Merge por QUANTIDADE
por_qtd = merge_sort(insumos, key=lambda x: x["quantidade"])
print("\nMergeSort por quantidade (crescente):")
for item in por_qtd:
    print(item["nome"], item["quantidade"])

# Quick por VALIDADE
qs_val = quick_sort(insumos, key=lambda x: (x["validade"]))
print("\nQuickSort por validade (próximos a vencer primeiro):")
for item in qs_val:
    print(item["nome"], item["validade"])
```
